## CONTEXTO

Una empresa educativa llamada X Education vende cursos en línea a profesionales del sector. Diariamente, muchos profesionales interesados ​​en los cursos visitan su sitio web y buscan información.

La empresa comercializa sus cursos en varios sitios web y motores de búsqueda como Google. Una vez que estas personas acceden al sitio web, pueden explorar los cursos, completar un formulario o ver videos. Al completar un formulario con su correo electrónico o número de teléfono, se les clasifica como clientes potenciales. Además, la empresa también obtiene clientes potenciales a través de referencias anteriores. Una vez adquiridos, los empleados del equipo de ventas comienzan a hacer llamadas, escribir correos electrónicos, etc. Mediante este proceso, algunos clientes potenciales se convierten, mientras que la mayoría no. La tasa de conversión típica de clientes potenciales en X Education ronda el 30 %.

Ahora bien, aunque X Education obtiene muchos clientes potenciales, su tasa de conversión es muy baja. Por ejemplo, si, por ejemplo, consiguen 100 clientes potenciales al día, solo unos 30 se convierten. Para que este proceso sea más eficiente, la empresa busca identificar los clientes potenciales con mayor potencial, también conocidos como "clientes potenciales calientes". Si identifican con éxito este grupo de clientes potenciales, la tasa de conversión debería aumentar, ya que el equipo de ventas se centrará más en comunicarse con ellos en lugar de llamar a todos.

Se generan muchos clientes potenciales en la etapa inicial (superior), pero solo unos pocos se convierten en clientes de pago en la etapa final. En la etapa intermedia, es necesario nutrir bien a los clientes potenciales (es decir, informarles sobre el producto, mantener una comunicación constante, etc.) para lograr una mayor conversión.

X Education busca seleccionar los clientes potenciales más prometedores, es decir, aquellos con mayor probabilidad de convertirse en clientes de pago. La empresa requiere que se cree un modelo en el que se asigne una puntuación a cada uno de ellos, de modo que los clientes cuntuaciónuación h más alta tengan una mayor probabilidad de conversión y los clientes puntuaciónpuuación h más baja, una menor. El director ejecutivo, en particular, ha estimado que la tasa de conversión objetivo de clientes potencidales rona el 80 %.

Vamos a construir un modelo de leadscoring, además de eso haremos una segmentación que nos permita entender mejor el tipo de leads que se están capturando y dar recomendaciones de consultoría.

Dataset: https://www.kaggle.com/datasets/amritachatterjee09/lead-scoring-dataset
