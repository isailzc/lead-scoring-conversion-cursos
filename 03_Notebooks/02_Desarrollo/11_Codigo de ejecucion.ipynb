```python
import cloudpickle
import pandas as pd
from janitor import clean_names

import warnings

warnings.filterwarnings("ignore", category=DeprecationWarning)

ruta_proyecto = 'C:/Users/Lenovo/Documents/portafolio_ds/automation/LEAD_KAGGLE'

nombre_fichero_datos = 'validacion.csv'

ruta_completa = ruta_proyecto + '/02_Datos/02_Validacion/' + nombre_fichero_datos

df = pd.read_csv(ruta_completa,index_col=0)

df = clean_names(df) 

df.rename(columns = {'totalvisits':'total_visits'}, inplace=True)

df = df.loc[(df.do_not_call !='Yes') & (df.do_not_email !='Yes') & (df.last_activity !='Email Bounced')] 
                     
variables_finales = ['asymmetrique_activity_score',
                    'last_notable_activity',
                    'lead_origin',
                    'page_views_per_visit',
                    'tags',
                    'total_time_spent_on_website',
                    'total_visits'
                  ]
                     
df = df[variables_finales]

nombre_pipe_ejecucion = 'pipe_ejecucion.pickle'

ruta_pipe_ejecucion = ruta_proyecto + '/04_Modelos/' + nombre_pipe_ejecucion

with open(ruta_pipe_ejecucion, mode='rb') as file:
   pipe_ejecucion = cloudpickle.load(file)

scoring = pipe_ejecucion.predict_proba(df)[:, 1]
```

Las siguientes celdas tienen el propósito de visualizar el scoring de cada registro. Lo importante es la parte de arriba.
